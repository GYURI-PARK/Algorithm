```python
# 2581 : 

m = int(input())
n = int(input())
res = []
sum = 0
for i in range(m, n+1):
    if i == 1: 
        continue
    for j in range(2, int(i** 0.5)+1 ):
        if i%j==0:
            break
    else:
        res.append(i)
        sum += i

if len(res) == 0:
  print(-1)

else :
  print(sum)
  print(res[0])
  

```

```python
# 5086 :  

while(True):
  m,n = map(int, input().split())
  if m < n and n % m == 0:
    print('factor')
  elif m > n and m % n == 0:
    print('multiple')
  elif m == 0 and n ==0:
    break;
  else:
    print('neither')
```
