```python
# 2609:  
# 1 try

x, y = map(int,input().split())
xli = []
yli = []
res1 = []
for i in range(2,x):
  if x % i == 0:
    xli.append(i)
for i in range(2,y):
  if y % i == 0:
    yli.append(i)
for i in range(len(xli)):
  if xli[i] in yli:
    res1.append(xli[i])
print(max(res1))
print((x//max(res1)) * (y//max(res1)) * max(res1))
```
