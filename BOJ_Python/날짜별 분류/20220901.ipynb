```python
# 17478 :  ?
# 

def recurFunction(i, n):
    print('-'*(i)*4,'" ?"')
    if i == n:
      print('_'*i*4,'"    "')
    else:
      print('_'*i*4,'" .          .')
      print('_'*i*4,'       ,    .')
      print('_'*i*4,'    .   ,      ."')
      recurFunction(i+1,n)
    print('_'*i*4,' .')

m = int(input())
print('       .')
recurFunction(0, m)
```

```python
# counter  
# 5    

from collections import Counter
n = int(input())
string_list = []
for i in range(n):
  string_list.append(input()[0])
cnt = Counter(string_list)
res = []
for i,j in cnt.items():
  if j >= 5:
    res.append(i)
if len(res) >= 1:
  print(''.join(sorted(res)))
else :
  print("PREDAJA")
```

```python
li = sorted([input()[0] for _ in range(int(input()))])
s = set(li)
res = []
for c in s:
    if li.count(c) >= 5:
        res.append(c)
print(''.join(sorted(res)) if len(res) > 0 else "PREDAJA")
```
