```python
# 1010번 : 다리놓기 
# math모듈 사용하지 않고 풀기

def factorial(n):
  res = 1
  for i in range(1,n+1):
    res *= i
  return res

n = int(input())
for i in range(n):
  n, m = map(int,input().split())
  case = factorial(m) // (factorial(n) * factorial(m - n))
  print(case)
```

```python
# 1194번 : 분수찾기

n = int(input())

line = 0
end = 0
while n > end:
    line += 1
    end += line

diff = end - n
if line%2 == 0: #짝수 라인 일때
    top = line - diff
    bottom = diff + 1
else:
    top = diff + 1
    bottom = line - diff

print("%d/%d"%(top,bottom))
```
