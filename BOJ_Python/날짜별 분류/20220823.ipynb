```python
# 1065번 : 한수

def hansu(num) :
    hansu_cnt = 0
    for i in range(1, num+1):
        num_list = list(map(int,str(i)))
        if i < 100:
            hansu_cnt += 1 
        elif num_list[0]-num_list[1] == num_list[1]-num_list[2]:
            hansu_cnt += 1 
    return hansu_cnt

num = int(input())
print(hansu(num))
```

```python
# 4948번 : 베르트랑 공준
# 시간초과

while(True):
    count = 0
    n = int(input())
    if n == 0:
        break;
    for i in range(n,2*n+1):
        for j in range(2, int(i**0.5) +1):
            if i%j == 0:
                break;
        else:
            count += 1
    print(count)
```

```python
# 10870번 : 피보나치 수5
# for 문으로 풀어보기

f0 = 0
f1 = 1
n = int(input())
if n == 0:
  print(f0)
elif n == 1:
  print(f1)
else:
  for i in range(n-1):
    f2 = f0 + f1
    f0 = f1
    f1 = f2
  print(f2)
```

```python
# 10870번 : 피보나치 수 5
# 재귀함수 사용하기

def fibonacci(num):
  if (num==0):
    res = 0
  elif(num==1):
    res = 1
  else:
    res = fibonacci(num-1) + fibonacci(num-2)

  return res
num = int(input())
print(fibonacci(num))
```
