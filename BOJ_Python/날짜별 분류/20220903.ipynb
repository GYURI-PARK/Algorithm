```python
# 9375 :  
#   +    

n = int(input())

for i in range(n):
  res = 1
  num = int(input())
  dic = {}
  #count = 1
  for i in range(num):
    thing, sort = input().split() 
    if sort in list(dic.keys()) :
      count = dic[sort]
      count += 1
      dic[sort] = count
    else :
      count = 1
      dic[sort] = count
  jli =[]
  for i,j in dic.items():
    jli.append(j)

  for i in range(len(jli)):
    res *= (jli[i] + 1)
  print(res-1)
```

```python
# 2447 :   -10

def star(num):
  for i in range(1,num+1):
    for j in range(1, num+1):
      if j%3 == 0:
        print('*')
      else:
        print('*', end='')
```

```python
def star2(num):
  for i in range(1,num+1):
    for j in range(1, num+1):
      print('*', end='')
      if j % 3 ==0:
        print('\n')
```

```python
def star3(num):
  array = [['*' for col in range(num)] for row in range(num)]
  for i in range(num//3):
    for j in range(num//3):
      array[num//2-i+int((num**(1/3))-1)][num//2-j+int((num**(1/3))-1)] = ' '
  for i in range(num):
    for j in range(num):
      print(array[i][j],end=' ')
    print('\n')
  
```
