```python
# 2004 :   0 
#  

n,k = map(int, input().split())
kfac = 1
nkfac = 1
# n! / k! * (n-k)!
count = 0

#   
for i in range(n-k,n+1):
    kfac *= i
for i in range(1, n-k+1):
    nkfac *= i

res = kfac // nkfac
for i in str(res)[::-1]:
  if i == '0':
    count +=1    
    continue;
  else:
    break;
print(count)
```

```python
n,k = map(int, input().split())
kfac = 1
nkfac = 1
count = 0
for i in range(1, n+1):
  if i < n-k:
    nkfac *= i
  elif i > n - k:
    kfac *= i
res = kfac // nkfac
for i in str(res)[::-1]:
  if i == '0':
    count +=1    
    continue;
  else:
    break;
print(count)
```
