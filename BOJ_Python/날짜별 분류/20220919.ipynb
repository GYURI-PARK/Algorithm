```python
# 1654번 : 랜선 자르기
# 반복문 + 함수 별도로 구현

k,n = map(int, input().split())
li = []
for i in range(k):
  li.append(int(input()))
li = sorted(li)

def binary(n, arr, cnt):
  start = 1
  end = max(arr)

  while start <= end :
    cnt = 0
    mid = (start + end) // 2
    for i in arr:
      cnt += (i // mid)
    
    if cnt >= n:
      start = mid + 1
    
    else:
      end = mid - 1

  return end

print(binary(n, li, 0))
```
