```python
# 이것도 쉽게 생각해보자
# 결국 모든 8 x 8 행렬 원소를 한 줄로 나열했을 때 앞에 있는 알파벳과 값이 달라야 한다.
# 그리고 n x m행렬이 주어질 때
# 행 (n-8), 열 (m-8) 먄큼 검사해서 
# 각각의 error 값 중 minimum을 출력하면 된다.

# 아니다 이렇게 하면 8*8 행렬에서 칠해야 하는 값을 찾지 못한다.

n, m = map(int, input().split())
# n:행, m:열

arr = [[0 for col in range(m)] for row in range(n)]
for i in range(n):
  arr[i] = input()

error = []
cnt = 0

li = []
for i in range(n):
  for j in range(m):
    li.append(arr[i][j])

# 행의 맨 마지막 알파벳과 그 다음 행 젤 첫번 째 알파벳은 같아야됨.

end_li = [7, 15,23, 31, 39, 47, 55, 63]

for i in range((n-8+1) + (m-8+1)):
  for j in range(i, 8*8-1):
    if li[j] == li[j+1] and j not in end_li:
      cnt += 1
      #print(j)
    elif j in end_li and li[j] != li[j+1] :
      cnt += 1
      #print(j)
  error.append(cnt-1)



print(cnt-1)
```

```python
dic = {}
li = []
for i in range(int(input())):
  kg, cm = input().split()
  li.append([int(kg), int(cm)])
  dic[i] = 1

for i in range(0, len(li)-1):
  for j in range(i+1, len(li)):
    if (li[i][0] > li[j][0]) and (li[i][1] > li[j][1]):
      dic[j] += 1
    elif (li[i][0] < li[j][0]) and (li[i][1] < li[j][1]):
      dic[i] += 1

for i,j in dic.items():
  print(j, end = ' ', sep = ' ')
```
