```python
# 1159번 : 농구 경기

# dictionary 형태로 (첫글자 : 횟수)
n = int(input())
sli = []
first_alphabet = {}

for i in range(n):
  sli.append(input()) 
  if sli[i][0] in first_alphabet:
    count = 1
    count += 1
    first_alphabet[sli[i][0]] = count
  else :
    first_alphabet[sli[i][0]] = 1
  

```

```python
from collections import Counter
n = int(input())
player = []
fn = []
cnt = 0
for i in range(n):
    a = input()
    player.append(a[0])
player_count = Counter(player)
for i, j in player_count.items():
    if j >= 5:
        fn.append(i)
        cnt += 1
fn.sort()
if cnt == 0:
    print("PREDAJA")
else:
    for i in fn:
        print(i, end='')
```

```python
li = sorted([input()[0] for _ in range(int(input()))])
s = set(li)
res = []
for c in s:
    if li.count(c) >= 5:
        res.append(c)
print(''.join(sorted(res)) if len(res) > 0 else "PREDAJA")
```

```python
x={}  
n = int(input()) 

for i in range(n):
    val = input()
    temp = val[0]
    if x.get(temp):
        x[temp]=x[temp]+1
    else : x[temp]=1

temp=[]  
index = 0  
for key,val in x.items():
    if int(val)>=5 :
        temp.insert(index, key)
        index+=1

temp.sort()
answer = ""   
for i in temp:
        answer+=i 
   
if answer=="": print("PREDAJA")
else : print(answer)
```
