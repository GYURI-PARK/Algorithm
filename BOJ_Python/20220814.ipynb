```python
# 9020번 : 골드바흐의 추측

# 5000번 까지의 소수 모두 구해서 list만들어 놓기

li = []
for i in range(2, 5001):
  for j in range(2,int(i**0.5)+1):
    if i % j == 0:
      break;
  else:
    li.append(i)

li
# 만약에 n이 주어진 짝수일 때
n = int(input())
for i in range(len(li)):
  li[i] = li[i] - n//2
  li[i] = abs(li[i])
```

```python

minIdx = li.index(min(li))
minIdx
```

```python
# 두 가지의 합이 n이 나오는 경우 먼저 뽑고
# 두 숫자의 차가 가장 작은 경울
n = int(input())
for a in li:
  for b in li:
    if (a+b == n):
      sub = abs(a-b)
      print(a,b)
      print(sub)
```
