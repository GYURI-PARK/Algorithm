```python
# 3036 : 

n = int(input())
li = []
res2 = []
li.append(input().split())
for i in range(1, n):
  res = str(int(li[0][0]) // int(li[0][i])) 
  if (int(li[0][0]) % int(li[0][i]) == 0):
    res += '/1'
    print(res)
  else:
    for j in range(2, min(int(li[0][0]), int(li[0][i]))+1):
      if int(li[0][0]) % j == 0 and int(li[0][i]) % j ==0 :
         res2.append(j)
    print(str(int(li[0][0])//max(res2)) + '/' + str(int(li[0][i])//max(res2)))
```
