```python
# 1065 : 

def hansu(num) :
    hansu_cnt = 0
    for i in range(1, num+1):
        num_list = list(map(int,str(i)))
        if i < 100:
            hansu_cnt += 1 
        elif num_list[0]-num_list[1] == num_list[1]-num_list[2]:
            hansu_cnt += 1 
    return hansu_cnt

num = int(input())
print(hansu(num))
```

```python
# 4948 :  
# 

while(True):
    count = 0
    n = int(input())
    if n == 0:
        break;
    for i in range(n,2*n+1):
        for j in range(2, int(i**0.5) +1):
            if i%j == 0:
                break;
        else:
            count += 1
    print(count)
```
