```python
# 2910번 : 빈도 정렬
# 자료 구조
# 정렬
# 해시를 사용한 집합과 맵
# 트리를 사용한 집합과 맵


# dictionary 사용
# 문제 n,c가 하는 역할이 없음..


n,c = map(int, input().split())
li = []
li.append(input().split())
li = li[0]
dic = {}
for i in li:
  if i not in dic:
    dic[i] = 1
  else:
    dic[i] += 1
dic = sorted(dic.items(), key=lambda x : -x[1])
for i in range(len(dic)):
  print((dic[i][0]+' ') * dic[i][1], end='')

# 맞았습니다
# n,c 살려서 다시 풀어보기ㅠㅠㅠㅠ
```
