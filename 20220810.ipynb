```python
# 4948번 : 베르트랑 공준 

while(True):
  count = 0
  n = int(input())
  if n == 0:
    break;
  elif n == 1:
    print(1)
  else:
    for i in range(n,2*n+1):
      for j in range(2, int(i**0.5)+1):
        if i%j == 0:
          break;
      else:
        count += 1
    print(count)
  
  
# !! 시간 초과 !!
  
```

```python
# 1037번 : 약수

n = int(input())
x = list(map(int,input().split()))
if n == 1:
  print(x[0]*x[0])

else:
  print(min(x) * max(x))
```

```python
# 2609번: 최대공약수와 최소공배수

x, y = map(int,input().split())
xli = []
yli = []
res1 = []
for i in range(2,x):
  if x % i == 0:
    xli.append(i)
for i in range(2,y):
  if y % i == 0:
    yli.append(i)
for i in range(len(xli)):
  if xli[i] in yli:
    res1.append(xli[i])
print(max(res1))
print((x//max(res1)) * (y//max(res1)) * max(res1))
```
