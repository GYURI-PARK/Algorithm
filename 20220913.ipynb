```python
# 1920번 : 수 찾기
# 이분 탐색 이용하기


n = int(input())
N = sorted(map(int, input().split()))
m = int(input())
M = map(int, input().split())

def binary(l, N, start, end):
    if start > end:
        return 0
    m = (start+end)//2
    if l == N[m]:
        return 1
    elif l < N[m]:
        return binary(l, N, start, m-1)
    else:
        return binary(l, N, m+1, end)

for l in M:
    start = 0
    end = len(N)-1
    print(binary(l,N,start,end))
```

```python
# 1259번 : 팰린드롬수

# s의 길이가 짝수일 때 팰린드롬수 여부 구하는 함수
def even(s):
  mid = len(s) // 2
  for i in range(len(s) // 2):
    if s[mid-i-1] != s[mid+i]:
      return 'no'
      break;
  return 'yes'

# s의 길이가 홀수일 때 팰린드롬수 여부 구하는 함수
def odd(s):
  mid = len(s) // 2
  for i in range(1, len(s) // 2 + 1):
    if s[mid-i] != s[mid+i]:
      return 'no'
      break;
  return 'yes'


while(True):
  s = input()
  if s == '0' :
    break;
  elif len(s) % 2 == 0:
    print(even(s))
  else:
    print(odd(s))
```
