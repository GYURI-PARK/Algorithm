```python
# 1920 :  
#   


n = int(input())
N = sorted(map(int, input().split()))
m = int(input())
M = map(int, input().split())

def binary(i, N, start, end):
    if start > end:
        return 0
    m = (start+end)//2
    if i == N[m]:
        return 1
    elif i < N[m]:
        return binary(i, N, start, m-1)
    else:
        return binary(i, N, m+1, end)

for i in M:
    start = 0
    end = len(N)-1
    print(binary(i,N,start,end))
```

```python
# 1259 : 

# s       
def even(s):
  mid = len(s) // 2
  for i in range(len(s) // 2):
    if s[mid-i-1] != s[mid+i]:
      return 'no'
      break;
  return 'yes'

# s       
def odd(s):
  mid = len(s) // 2
  for i in range(1, len(s) // 2 + 1):
    if s[mid-i] != s[mid+i]:
      return 'no'
      break;
  return 'yes'


while(True):
  s = input()
  if s == '0' :
    break;
  elif len(s) % 2 == 0:
    print(even(s))
  else:
    print(odd(s))
```
