```python
# 9020 :  

# 5000     list 

li = []
for i in range(2, 5001):
  for j in range(2,int(i**0.5)+1):
    if i % j == 0:
      break;
  else:
    li.append(i)

li
#  n   
n = int(input())
for i in range(len(li)):
  li[i] = li[i] - n//2
  li[i] = abs(li[i])
```
