```python
# 11650번 : 좌표 정렬하기

n = int(input())
li = []
for i in range(n):
  li.append(list(map(int, input().split())))
li.sort(key = lambda x :(x[0], x[1]))
for i in li:
  print(i[0],i[1])

```

```python
# 1181번 : 단어 정렬

n = int(input())
sli = []
for i in range(n):
  sli.append(input())
res = list(set(sli))
res.sort(key = lambda x : (len(x), x))
for x in res:
  print(x)
    
```
