```python
# 18870번 : 좌표 압축
# 시간 초과

n = int(input())
li = []
li.append(list(input().split()))
res = list(set(li[0]))
res.sort(key = lambda x:x)
for i in li[0]:
  for j in range(len(res)):
    if i == res[j]:
      print(j, sep=' ', end=' ')
```

```python
# 이거도 시간초과...

n = int(input())
li_dic = {}
li = []
li.append(list(input().split()))
res = list(set(li[0]))
res.sort(key = lambda x:x)
# list to string
strres = "".join(res)
for i in li[0]:
  li_dic[i] = strres.find(i)
  print(li_dic[i], sep=' ', end = ' ')

# (li_dic[i], sep=' ', end = ' ') 띄어쓰기 
```

```python
n = int(input())
li_dic = {}
li = []
li.append(list(input().split()))
res = list(set(li[0]))
res.sort(key = lambda x:int(x))
for i in range(len(res)):
  li_dic[res[i]] = i
for i in li[0]:
  print(li_dic[i], sep=' ', end = ' ')
```
