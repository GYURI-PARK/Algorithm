```python
sen = str(input().upper())
freq = []
for char in set(sen):
    count = sen.count(char)
    freq.append([count, char])
freq.sort(reverse = True)

if (len(freq)>1 and freq[0][0] == freq[1][0]):
    print("?")
else:
    print(freq[0][1])
```
