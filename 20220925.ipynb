```python
# 2164 :  2
#   (RecursionError)

n = int(input())
queue = [i for i in range(1, n+1)]

def card2(n, queue):
  while(len(queue) > 1):
    if n == 1:
      queue.pop(0)
      card2(0, queue)
    else:
      a = queue.pop(0)
      queue.append(a)
      card2(1, queue)
    
  return queue[0]
print(card2(n,queue))
```

```python
# 

n = int(input())
queue = [i for i in range(1, n+1)]
cnt = 1
while(len(queue) > 1):
  if cnt % 2 == 1:
    queue.pop(0)
    cnt += 1

  else:
    a = queue.pop(0)
    queue.append(a)
    cnt += 1


print(queue[0])
```

```python
# ! 

from collections import deque

n = int(input())
queue = deque(i for i in range(1, n+1))
cnt = 1
while(len(queue) > 1):
  if cnt % 2 == 1:
    queue.popleft()
    cnt += 1


  else:
    a = queue.popleft()
    queue.append(a)
    cnt += 1


print(queue[0])
```
