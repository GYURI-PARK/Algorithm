```python
# 10828 : 

n = int(input())
li = []
for i in range(n):
  order = input().split()
  if 'push' in order :
    li.append(order[1])
  elif order[0] == 'top':
    if len(li) != 0 :
      print(li[len(li)-1])
    else:
      print(-1)
  elif order[0] == 'size':
    print(len(li))
  elif order[0] == 'empty':
    if len(li) == 0:
      print(1)
    else:
      print(0)
  else :
    if len(li) == 0:
      print(-1)
    else:
      print(li.pop())




```
