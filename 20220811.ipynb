```python
# 2609번: 최대공약수와 최소공배수
# 1 try

x, y = map(int,input().split())
xli = []
yli = []
res1 = []
for i in range(2,x):
  if x % i == 0:
    xli.append(i)
for i in range(2,y):
  if y % i == 0:
    yli.append(i)
for i in range(len(xli)):
  if xli[i] in yli:
    res1.append(xli[i])
print(max(res1))
print((x//max(res1)) * (y//max(res1)) * max(res1))
```

```python
# 25304번 : 영수증

sum = int(input())
n = int(input())
res = 0
for i in range(n):
  x,y = map(int, input().split())
  res += x*y
if res == sum:
  print("Yes")
else:
  print("No")
```

```python
# 9020번: 골드바흐의 추측

# 5000번 까지의 소수 모두 구해서 list만들어 놓기

li = []
for i in range(2, 5001):
  for j in range(2,int(i**0.5)+1):
    if i % j == 0:
      break;
  else:
    li.append(i)

li
# 만약에 n이 주어진 짝수일 때
n = int(input())
for i in range(len(li)):
  li[i] = li[i] - n//2
  li[i] = abs(li[i])
```

```python
# 두 가지의 합이 n이 나오는 경우 먼저 뽑고
# 두 숫자의 차가 가장 작은 경울
n = int(input())
for a in li:
  for b in li:
    if (a+b == n):
      sub = abs(a-b)
      print(a,b)
      print(sub)
```

```python
# 1934번 : 최소공배수

n = int(input())
for i in range(n):
  x,y = map(int,input().split())
  if x == 1 or y == 1:
    print(max(x,y))
  else:
    res = []
    for i in range(1, min(x,y)+1):
      if x%i == 0 and y%i == 0:
        res.append(i)
    print((x//max(res)) * (y//max(res)) * max(res))
```
