```python
# 7568번 : 덩치

dic = {}
li = []
for i in range(int(input())):
  kg, cm = input().split()
  li.append([int(kg), int(cm)])
  dic[i] = 1

for i in range(0, len(li)-1):
  for j in range(i+1, len(li)):
    if (li[i][0] > li[j][0]) and (li[i][1] > li[j][1]):
      dic[j] += 1
    elif (li[i][0] < li[j][0]) and (li[i][1] < li[j][1]):
      dic[i] += 1

for i,j in dic.items():
  print(j, end = ' ', sep = ' ')
```

```python
# 1436번 : 영화감독 숌

num1 = '666'
plus = 1
n = int(input())
for i in range(0,n-1):
  for j in range(i+1, n):
    if int(num1 + str(plus)) > int(str(plus) + num1):
      num =  str(plus) + num1
      plus += 1
      print(num)

```

```python
# 1018번 : 체스판 다시 칠하기
# 틀렸다
# 문제 제대로 안읽음
# 8*8 최소 에러값 찾기

n, m = map(int, input().split())
# n:행, m:열

arr = [[0 for col in range(m)] for row in range(n)]
for i in range(n):
  arr[i] = input()

# 맨 왼쪽 위에 바둑돌이 W일 때와 B일 때로 나누기

# 정답 array 만들어 놓기
array = [[0 for col in range(m)]for row in range(n)]
if arr[0][0] == 'W':
  for i in range(n):
    for j in range(m):
      if (i+j) % 2 == 0:
        array[i][j] = 'W'
      else:
        array[i][j] = 'B'

else:
  for i in range(n):
    for j in range(m):
      if (i+j) % 2 == 0:
        array[i][j] = 'B'
      else:
        array[i][j] = 'W'

error = 0

# ====================
# 비교해서 다르면 error += 1
for i in range(n):
  for j in range(m):
    if arr[i][j] != array[i][j]:
      error += 1
      #print(i,j)


print(error)

```
