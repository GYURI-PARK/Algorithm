```python
# 2002번 : 추월

n = int(input())
in_order = {}
out_order = {}
for i in range(n):
  in_order[input()] = i+1

for i in range(n):
  out_order[input()] = i+1

# in_order 의 첫번째 값보다 앞에있는 차들 뽑기
# out_order에서의 순서값 찾기
firt = 0
cnt = 0
for i,j in in_order.items():
  if j == 1:
    first = out_order[i]
  for m,n in out_order.items():
    if n > first and i == m and n < j :
      print(cnt)
      cnt += 1


cnt += first - 1
# 일단 first 값 - 1명이 추월 한거임.
# 젤 먼저 나와야 되는 차보다 먼저 나옴
print(cnt)
```
