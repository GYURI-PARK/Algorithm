```python
# 4948번 : 베르트랑 공준 

count = 0
res = 0
while(True):
  n = int(input())
  for i in range(n, 2*n+1):
    for j in range(2, int(i** 0.5)+1):
      if i%j == 0:
        break;
      else: 
        count += 1
  if count > 0 :
    res += 1
  print(res)
  count = 0
  res = 0
  if n == 0:
    break;
```
