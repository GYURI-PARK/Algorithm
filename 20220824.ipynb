```python
# 2775번 : 부녀회장이 될테야
# 제한 1 ≤ k, n ≤ 14 
# -> 14x14만들어놓고 시작

array = [[0 for col in range(14)] for row in range(15)]
for i in range(15):
  for j in range(14):
    array[0][j] = j+1
    array[i][0] = 1
    if i > 0 and j > 0:
      array[i][j] = array[i-1][j]+ array[i][j-1]
t = int(input())
for i in range(t):
  k = int(input())
  n = int(input())
  print(array[k][n-1])    
```

```python
# 1676번 : 팩토리얼 0의 개수
# 문자열을 뒤에서 부터 판단하는 방법


n = int(input())
fac = 1
count = 0
for i in range(1,n+1):
  fac *= i
for j in range(1,len(str(fac))+1):
  if list(str(fac))[len(str(fac))-j] == '0':
    count += 1
    continue;
  else:
    break;
print(count)
```

```python
# 2004번 : 조합  0의 개수
# 시간 초과

n,k = map(int, input().split())
kfac = 1
nkfac = 1
# n! / k! * (n-k)!
count = 0

# 반복문 두개 합치기
for i in range(n-k,n+1):
    kfac *= i
for i in range(1, n-k+1):
    nkfac *= i

res = kfac // nkfac
for i in str(res)[::-1]:
  if i == '0':
    count +=1    
    continue;
  else:
    break;
print(count)
```

```python
n,k = map(int, input().split())
kfac = 1
nkfac = 1
count = 0
for i in range(1, n+1):
  if i < n-k:
    nkfac *= i
  elif i > n - k:
    kfac *= i
res = kfac // nkfac
for i in str(res)[::-1]:
  if i == '0':
    count +=1    
    continue;
  else:
    break;
print(count)
```
