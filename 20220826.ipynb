```python
# 2004번 : 조합 0의 개수
# 시간초과

n,k = map(int, input().split())
kfac = 1
nkfac = 1
count = 0
for i in range(1, n+1):
  if i < n-k:
    nkfac *= i
  elif i > n - k:
    kfac *= i
res = kfac // nkfac

for j in range(1,len(str(res))+1):
  if list(str(res))[len(str(res))-j] == '0':
    count += 1
    continue;
  else:
    break;
print(count)
```
