```python
# 2805번 : 나무 자르기

N, M = map(int, input().split())
trees = list(map(int, input().split()))
trees.sort()

def binary(M, trees):
    start = 1
    end = trees[-1]
    while start <= end:
        cuts = 0
        mid = (start + end) // 2
        for tree in trees:
            if mid < tree:
                cuts += tree - mid
        if cuts == M:
            return mid
        elif cuts < M:
            end = mid - 1
        else:
            start = mid + 1
    return end

print(binary(M, trees))
```

```python
# 반복문 사용한 이진탐색 코드

def binary_search(array, target):

  start = 0
  end = len(array) - 1
  
  while start <= end:

    mid = (start + end) // 2
    # 원하는 값 찾은 경우 인덱스 반환
    if array[mid] == target:
      return mid

    # 원하는 값이 mid보다 작은 경우 왼쪽으로 범위 이동
    elif array[mid] > target:
      end = mid - 1


    else:
      start = mid + 1

    return None
```
