```python
# 1010 :  
# math   

def factorial(n):
  res = 1
  for i in range(1,n+1):
    res *= i
  return res

n = int(input())
for i in range(n):
  n, m = map(int,input().split())
  case = factorial(m) // (factorial(n) * factorial(m - n))
  print(case)
```
