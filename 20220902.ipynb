```python
# 14425번 : 문자열 집합
# 틀렸습니다
# n에 중복되는 문자열이 들어가는 경우 각각 카운팅 되어야 하는데 set형식이므로 하나로 취급된다.

m, n = map(int, input().split())
mli = []
nli = []
for i in range(m):
    mli.append(input())
for i in range(n):
    nli.append(input())
print(len(set(mli) & set(nli)))
```

```python
# 성공
# 시간 너무 많이 걸림

m, n = map(int, input().split())
mli = []
nli = []
for i in range(m):
    mli.append(input())
for i in range(n):
    nli.append(input())

res = list(set(mli) & set(nli))
cnt = 0
for i in nli:
  if i in res:
    cnt += 1
print(cnt)
```

```python
n, m=map(int, input().split())
s=set()
cnt=0
for _ in range(n):
    s.add(input())
for i in range(m):
    if input() in s:
        cnt+=1
print(cnt)
```
