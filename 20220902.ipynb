```python
# 14425 :  
# 
# n         set  .

m, n = map(int, input().split())
mli = []
nli = []
for i in range(m):
    mli.append(input())
for i in range(n):
    nli.append(input())
print(len(set(mli) & set(nli)))
```

```python
# 
#    

m, n = map(int, input().split())
mli = []
nli = []
for i in range(m):
    mli.append(input())
for i in range(n):
    nli.append(input())

res = list(set(mli) & set(nli))
cnt = 0
for i in nli:
  if i in res:
    cnt += 1
print(cnt)
```

```python
n, m=map(int, input().split())
s=set()
cnt=0
for _ in range(n):
    s.add(input())
for i in range(m):
    if input() in s:
        cnt+=1
print(cnt)
```
