```python
# 1769번 : 3의 배수

# num == str로 입력받기

def of3(i, num):
  if len(num) == 1:
    if int(num) % 3 == 0:
      print(i)
      print("YES")
    else:
      print(i)
      print("NO")
  else:
    i += 1
    new_num = 0
    for j in range(len(num)):
      new_num += int(num[j])
    of3(i, str(new_num))
of3(0, input())
```
