```python
# 10816번 : 숫자 카드 2
# 시간 초과

from collections import Counter

n = int(input())
nli = input().split()

m = int(input())
mli = input().split()


res_dic = {}
for i,j in Counter(nli).items():
  if i in mli:
    res_dic[i] = j
for i in mli:
  if i in res_dic.keys():
    print(res_dic[i], sep=' ', end=' ')
  else:
    print(0, sep=' ', end=' ')
```

```python
# 10816번 : 숫자 카드 2
# 성공

from collections import Counter

n = int(input())
nli = input().split()
dic = {}
for i in range(n):
  if nli[i] in dic.keys():
    dic[nli[i]] += 1
  else:
    dic[nli[i]] = 1
m = int(input())
mli = input().split()
for i in mli:
  if i in dic.keys():
    print(dic[i], sep=' ', end=' ')
  else:
    print(0, sep=' ', end=' ')
```
