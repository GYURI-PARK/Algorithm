```python
#   
#   8 x 8            .
#  n x m  
#  (n-8),  (m-8)   
#  error   minimum  .

#    8*8      .

n, m = map(int, input().split())
# n:, m:

arr = [[0 for col in range(m)] for row in range(n)]
for i in range(n):
  arr[i] = input()

error = []
cnt = 0

li = []
for i in range(n):
  for j in range(m):
    li.append(arr[i][j])

#            .

end_li = [7, 15,23, 31, 39, 47, 55, 63]

for i in range((n-8+1) + (m-8+1)):
  for j in range(i, 8*8-1):
    if li[j] == li[j+1] and j not in end_li:
      cnt += 1
      #print(j)
    elif j in end_li and li[j] != li[j+1] :
      cnt += 1
      #print(j)
  error.append(cnt-1)



print(cnt-1)
```

```python
dic = {}
li = []
for i in range(int(input())):
  kg, cm = input().split()
  li.append([int(kg), int(cm)])
  dic[i] = 1

for i in range(0, len(li)-1):
  for j in range(i+1, len(li)):
    if (li[i][0] > li[j][0]) and (li[i][1] > li[j][1]):
      dic[j] += 1
    elif (li[i][0] < li[j][0]) and (li[i][1] < li[j][1]):
      dic[i] += 1

for i,j in dic.items():
  print(j, end = ' ', sep = ' ')
```
