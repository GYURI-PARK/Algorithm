```python
# 1436번 : 영화감독 숌

# 숫자를 입력받으면 입력 받은 숫자 만큼 돌아간다.
# prevnum =  666
# cnt = 1
# i = 1부터 더해지고, 
# while cnt = int(input())
# int('666' + 'i') > int('i' + '666') and int('i' + '666') > prevnum => nextnum = int('i' + '666')
# cnt += 1
# i에 6이 들어갈 경우 = > prevnum에서 666앞에 숫자 하나가 빠지고 뒤에 + 0부터 9까지

# 7 이상부터 시간 너무 많이 걸림

n = int(input())
prevnum = '666'
cnt = 1
i = 1
while(cnt < n):
  if(int('666'+str(i)) > int(str(i)+'666') and int(str(i) + '666') > int(prevnum)):
    nextnum = int(str(i)+'666')
    cnt += 1
    prevnum = nextnum
    i += 1
    if '6' in 'i':
      for j in range(0,10):
        start_idx = str(prevnum).find('666')
        nextnum = prevnum[start_idx:]+str(j)
        prevnum = nextnum
print(prevnum)
```
