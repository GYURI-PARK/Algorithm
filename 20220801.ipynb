```python
# 2869번 : 달팽이는 올라가고 싶다.

a, b, v = map(int,input().split())
cur = 0
count = 0
while (True):
  cur += a
  count += 1
  if cur > v:
    count -= 1
    break;
  else:
    cur -= b
  
print(count)
```

```python
# 1193번 : 분수찾기

# 대각선 대로 분자 분모 합이 같다

1 - 2 - 3 - 4 - 5
1 - 4 - 9 - 16 - 25 
1 - 3 - 6 - 10 - 15 - 21 - 28 
# 앞 = 포함 x 뒤 = 포함

# 먼저 합이 얼마가 되는지 찾아서 그 안에서 다시 규칙찾기

n = int(input())
count = 0
i = 0
sum = 0
while (sum < 30):
  sum += i
  #print(sum)
  i += 1
  if n > sum and (n <= (i+1)*(i+1) - sum) :
    print(i+1)
    break;
```

```python
# 1427번 : 소트인사이드

# 선택 정렬
# 오름차순
s = list(input())
for i in range(len(s)-1):
  minIdx = i
  for j in range(i+1, len(s)):
    if s[j] < s[minIdx]:
      minIdx = j
  temp = s[minIdx]
  s[minIdx] = s[i]
  s[i] = temp

print(s)
```

```python
# 선택정렬
# 내림차순

s = list(input())
for i in range(len(s)-1):
  maxIdx = i
  for j in range(i+1, len(s)):
    if s[j] > s[maxIdx]:
      maxIdx = j
  temp = s[maxIdx]
  s[maxIdx] = s[i]
  s[i] = temp

for i in range(len(s)):
  print(s[i], end = '')
```

```python
# 버블정렬
# 오름차순

s = list(input())
for i in range(len(s)-1,0,-1):
  for j in range(i):
    if s[j] < s[j+1]:
      temp = s[j]
      s[j] = s[j+1]
      s[j+1] = temp

for i in range(len(s)):
  print(s[i], end = '')
```
