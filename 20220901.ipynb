```python
# counter 사용해서 풀기
# 5개 넘는 첫글자 모두 출력

from collections import Counter
n = int(input())
string_list = []
for i in range(n):
  string_list.append(input()[0])
cnt = Counter(string_list)
res = []
for i,j in cnt.items():
  if j >= 5:
    res.append(i)
if len(res) >= 1:
  print(''.join(sorted(res)))
else :
  print("PREDAJA")
```

```python
li = sorted([input()[0] for _ in range(int(input()))])
s = set(li)
res = []
for c in s:
    if li.count(c) >= 5:
        res.append(c)
print(''.join(sorted(res)) if len(res) > 0 else "PREDAJA")
```
