```python
# 17478 :  ?
# 

def recurFunction(i, n):
    print('-'*(i)*4,'" ?"')
    if i == n:
      print('_'*i*4,'"    "')
    else:
      print('_'*i*4,'" .          .')
      print('_'*i*4,'       ,    .')
      print('_'*i*4,'    .   ,      ."')
      recurFunction(i+1,n)
    print('_'*i*4,' .')

m = int(input())
print('       .')
recurFunction(0, m)
```
