```python
# 1764번 : 듣보잡
# 시간초과

n,m = map(int,input().split())
li = []
for i in range(n+m):
  li.append(input())

res = set(li[0:n]) & set(li[n:n+m])
print(len(res))
for i in range(len(res)):
  print(sorted(list(res))[i])
```

```python
# 시간초과

n,m = map(int,input().split())
li = []
cnt = 0
res = []
for i in range(n+m):
  a = input()
  if a in li:
    cnt += 1
    res.append(a)
  li.append(a)
print(cnt)
for i in range(cnt):
  print(sorted(res)[i])
```

```python
# 1764번 : 듣보잡
# 출력형식이 잘못되었습니다
# 시간초과

n,m = map(int,input().split())
li1 = []
li2 = []

for i in range(n):
  li1.append(input())
for i in range(m):
  li2.append(input())
print(len(set(li1) & set(li2)))
for i in range(len(set(li1) & set(li2))):
  print(sorted(set(li1) & set(li2))[i])
```

```python
# 1764번 : 듣보잡
# 성공

n,m = map(int,input().split())
li1 = set()
li2 = set()
import sys

for i in range(n):
    li1.add(sys.stdin.readline().rstrip())
for i in range(m):
    li2.add(sys.stdin.readline().rstrip())
print(len(li1 & li2))
sort = sorted(li1 & li2)
for i in range(len(sort)):
    print(sort[i])
```
