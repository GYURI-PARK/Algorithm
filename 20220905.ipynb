```python
# 2231 : 
#      

n = int(input())
res = []
for i in reversed(range(n)):
  sum = 0
  sum += i
  for j in range(len(str(i))):
    sum += int(str(i)[j])
  #print(sum)
  if sum == n :
    res.append(i)

if len(res) == 0:
  print(0)
else:
  print(res[len(res)-1])

```

```python
# num = int
def maker(num):
  res = num
  for i in range(len(str(num))):
    res += int(str(num)[i])
  return(num, res)

li = []
for i in range(1,1000000):
  li.append(maker(i))


n = int(input())
if n in li:
  print(maker(n))
```
