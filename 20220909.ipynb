```python
# 10994번 : 별 찍기 - 19

def star(num) :
  arr = [[0 for col in range(num)]for row in range(num)]
  if num == 1:
    arr[0][0] = str('*')
  return arr

star(1)
```

```python
# 2501번 : 약수 구하기
# 브루트 포스 알고리즘 


n, k = map(int, input().split())

# n의 약수 구하기
li = []
for i in range(1, n+1):
  if n % i == 0:
    li.append(i)
if k > len(li):
  print(0)
else:
  print(li[k-1])
```
