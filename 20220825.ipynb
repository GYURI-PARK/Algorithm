```python
# 4948번 : 베르트랑 공준
# 시간  초과

while(True):
  n = int(input())
  res = []
  if n == 0:
    break;
  else:
    for i in range(n+1, 2*n+1):
      for j in range(2,int(i** 0.5)+1):
        if i % j == 0:
          break;
      else :
        res.append(i)
  print(len(res))
```

```python
# 9020번 : 골드바흐의 추측

# 2에서 5000사이의 소수 list 만들어 놓기

def isPrime(num):
  if num == 1:
    return False
  else:
    for i in range(2, int(num**0.5) +1):
      if num % i ==0:
        return False
    return True
primeLi = []
for i in range(2,5000):
  if isPrime(i):
    primeLi.append(i)

# primeLi 안에 있는 두 개의 숫자 합이 num이 되는 수 고르기 
res = []
n = int(input())
for i in range(n):
  num = int(input())
for i in primeLi:
  for j in primeLi:
    if i + j == num:
      res.append([i,j])
# res[전체 길이의 2분의 1만 돌리가] 리스트 안에서 차이가 최소값 출력
index = 0
for i in res[0:len(res)//2]:
  min = 0
  if abs(i[0] - i[1]) >= min :
    min = (abs(i[0] - i[1]))
    index += 1
print(res[index-1][0], res[index-1][1])
```

```python
# 숫자 하나 입력할 때 마다 결과값 나오게 출력
# 시간 초과

def isPrime(num):
  if num == 1:
    return False
  else:
    for i in range(2, int(num**0.5) +1):
      if num % i ==0:
        return False
    return True
primeLi = []
for i in range(2,5000):
  if isPrime(i):
    primeLi.append(i)

n = int(input())
for i in range(n):
  num = int(input())
  res = []
  for i in primeLi:
    for j in primeLi:
      if i + j == num:
        res.append([i,j])
  index = 0
  for i in res[0:int(round(len(res)/2))]:
    min = 0
    if abs(i[0] - i[1]) >= min :
      min = (abs(i[0] - i[1]))
      index += 1
  print(res[index-1][0], res[index-1][1])
```
