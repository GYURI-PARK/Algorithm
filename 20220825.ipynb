```python
#        
#  

def isPrime(num):
  if num == 1:
    return False
  else:
    for i in range(2, int(num**0.5) +1):
      if num % i ==0:
        return False
    return True
primeLi = []
for i in range(2,5000):
  if isPrime(i):
    primeLi.append(i)

n = int(input())
for i in range(n):
  num = int(input())
  res = []
  for i in primeLi:
    for j in primeLi:
      if i + j == num:
        res.append([i,j])
  index = 0
  for i in res[0:int(round(len(res)/2))]:
    min = 0
    if abs(i[0] - i[1]) >= min :
      min = (abs(i[0] - i[1]))
      index += 1
  print(res[index-1][0], res[index-1][1])
```
